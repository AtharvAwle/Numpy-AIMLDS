## Practicing Numpy with real world data

```python

import numpy as np
import matplotlib.pyplot as plt
```

```python
# Sales data of different foof joints over 4 years

sales_data = np.array([
    [1,150000,180000,220000,250000],   #Paradise Biryani
    [2,120000,140000,160000,190000],   #Spicy Villa
    [3,200000,230000,260000,300000],   #Tasty Tandoor
    [4,180000,210000,240000,270000],   #Burger King
    [5,160000,185000,205000,230000]    #Pizza Palace
])

print("=== Zomato Sales Analysis ===")
print("Shape of the dataset: ", sales_data.shape)

# Get the sales data of first three fast food joints over 4 years
print("Sales data of first 3 elements: ", sales_data[:3])


# Total Sales per year
yearly_Total = np.sum(sales_data[:, 1:],axis=0)
print("The year wise total sales is: ", yearly_Total)


# Minimum sales per restraunt
min_sales = np.min(sales_data[:,1:], axis=1)
print("The minimum sales for each restraunt is: ",min_sales)

# Maximum sales per year
max_sales_year = np.max(sales_data[:,1:],axis=0)
print("The max sales per year across zomato is: ", max_sales_year)

# Average sales per restraunt
avg_sales_rest = np.average(sales_data[:,1:],axis=1)
print("The average sales per restraunt is: ", avg_sales_rest)
```

```python
cumsum = np.cumsum(sales_data[:,1:],axis=1)

# Ploting example using matplotlib
plt.Figure(figsize=(8,6))
plt.plot(np.mean(cumsum,axis=0))
plt.title("Average cumulative sales across all restraunt in years")
plt.xlabel("YEARS")
plt.ylabel("SALES")
plt.grid(True)
plt.show()
```
